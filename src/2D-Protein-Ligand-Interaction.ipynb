```python
import os

folder = 'Prot-ligand-complexes'

def is_lig_present(lines):
    
    import re
    pattern = re.compile('^HETATM')
    
    for line in lines:
        if pattern.match(line):
            tokens = re.split("\s{1,}", line)
            if tokens[3] == 'LIG':
                return True
    return False

is_all_good = True 
for file in os.listdir(folder):
    abs_file_path = os.path.join(folder, file)
    
    with open(abs_file_path) as file:
        lines = file.readlines()
        
        if not is_lig_present(lines):
            is_all_good = False
            print(file)
            
if is_all_good:
    print('All contains ligands')
```
